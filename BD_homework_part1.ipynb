# Задача

Один из способов повысить эффективность взаимодействия банка с клиентами — отправлять предложение о новой услуге не всем клиентам, а только некоторым, которые выбираются по принципу наибольшей склонности к отклику на это предложение.

Задача заключается в том, чтобы предложить алгоритм, который будет выдавать склонность клиента к положительному или отрицательному отклику на предложение банка. Предполагается, что, получив такие оценки для некоторого множества клиентов, банк обратится с предложением только к тем, от кого ожидается положительный отклик.


Для решения этой задачи загрузите файлы из базы в Postgres.
Эта БД хранит информацию о клиентах банка и их персональные данные, такие как пол, количество детей и другие.

Описание таблиц с данными представлено ниже.

Описание статусов относительно работы:

**D_salary**

Описание информации о заработной плате клиентов:

**D_loan**

Информация о кредитной истории клиента:
- ID_CLIENT — идентификатор клиента;

**D_close_loan**

Информация о статусах кредита (ссуд):

## Задание 1

Соберите всю информацию о клиентах в одну таблицу, где одна строчка соответствует полной информации об одном клиенте.

```python
!git clone https://github.com/aiedu-courses/stepik_linear_models.git
```

```python
import os
import pandas as pd

data_folder = '/content/stepik_linear_models/datasets'

# Список файлов данных
data_files = os.listdir('/content/stepik_linear_models/datasets')

# Создание словаря для хранения DataFrame'ов
dataframes = {}

# Загрузка каждого файла в DataFrame и сохранение в словаре
for file in data_files:
    file_path = os.path.join(data_folder, file)
    dataframes[file] = pd.read_csv(file_path)

# Вывод имен файлов и первых нескольких строк каждого DataFrame
for file, df in dataframes.items():
    print(f"DataFrame from {file}")
```

```python

# Renaming the 'ID' column to 'ID_CLIENT' in the D_clients.csv dataset
dataframes['D_clients.csv'].rename(columns={'ID': 'ID_CLIENT'}, inplace=True)

# Repeating the merge process with the corrected column name
# Merging D_clients.csv with D_salary.csv on ID_CLIENT
merged_df = pd.merge(dataframes['D_clients.csv'], dataframes['D_salary.csv'], on='ID_CLIENT', how='left')

# Merging with D_target.csv on ID_CLIENT
merged_df = pd.merge(merged_df, dataframes['D_target.csv'], on='ID_CLIENT', how='left')

# Calculating LOAN_NUM_TOTAL (total number of loans for each client)
loan_count = dataframes['D_loan.csv'].groupby('ID_CLIENT').size().reset_index(name='LOAN_NUM_TOTAL')

# Merging with loan_count
merged_df = pd.merge(merged_df, loan_count, on='ID_CLIENT', how='left')

# Calculating LOAN_NUM_CLOSED (total number of closed loans for each client)
closed_loan_count = dataframes['D_close_loan.csv'][dataframes['D_close_loan.csv']['CLOSED_FL'] == 1].groupby('ID_LOAN').size().reset_index(name='LOAN_NUM_CLOSED')

# Merging D_loan.csv with closed_loan_count to map ID_LOAN to ID_CLIENT
closed_loans_per_client = pd.merge(dataframes['D_loan.csv'], closed_loan_count, on='ID_LOAN', how='left')

# Summing up closed loans per client
closed_loans_per_client = closed_loans_per_client.groupby('ID_CLIENT')['LOAN_NUM_CLOSED'].sum().reset_index()

# Merging with closed_loans_per_client
merged_df = pd.merge(merged_df, closed_loans_per_client, on='ID_CLIENT', how='left')

# Selecting only the required columns
final_columns = ['AGREEMENT_RK', 'TARGET', 'AGE', 'SOCSTATUS_WORK_FL', 'SOCSTATUS_PENS_FL', 'GENDER', 'CHILD_TOTAL',
                 'DEPENDANTS', 'PERSONAL_INCOME', 'LOAN_NUM_TOTAL', 'LOAN_NUM_CLOSED']
final_dataset = merged_df[final_columns]

# Displaying the first few rows of the final dataset
final_dataset.head()



```

**Удаление дубликатов.**
*   Оценка и обработка пропущенных значений. В зависимости от контекста и количества пропущенных значений, их можно удалить

```python
# Checking for duplicates in the final dataset
duplicates = final_dataset.duplicated().sum()

# Checking for missing values in key columns
missing_values = final_dataset.isnull().sum()

# Overview of duplicates and missing values
duplicates, missing_values
```

Дубликаты успешно удалены из датасета. Теперь давайте перейдем к обработке пропущенных значений.

На основе количества пропущенных значений в каждом столбце, мы можем:

*   Удалить строки с пропущенными значениями в столбцах `AGREEMENT_RK`, `TARGET`, `PERSONAL_INCOME`, `LOAN_NUM_TOTAL`, и `LOAN_NUM_CLOSED`, так как они являются ключевыми для анализа и их отсутствие может существенно повлиять на результаты.


```python
# Removing duplicates
final_dataset_cleaned = final_dataset.drop_duplicates()

# Re-checking for duplicates
duplicates_after_removal = final_dataset_cleaned.duplicated().sum()

# Re-checking for missing values
missing_values_after_removal = final_dataset_cleaned.isnull().sum()

duplicates_after_removal, missing_values_after_removal

```

Теперь датасет очищен от дубликатов и пропущенных значений, что делает его более подходящим для анализа и моделирования.

```python
# Removing rows with missing values in key columns
final_dataset_cleaned = final_dataset_cleaned.dropna(subset=['AGREEMENT_RK', 'TARGET', 'PERSONAL_INCOME', 'LOAN_NUM_TOTAL', 'LOAN_NUM_CLOSED'])

# Checking the shape of the dataset after removal of missing values
final_dataset_cleaned.shape

```
